```python
#The Libraries are installed for following purpose
#panadas for dataframes
#matplotlib for plotting/data visualization 
#seaborn for boxplot,heatmaps and distance plot
#numpy for log and random function
#stats for probability plots: probplot()

#requests library for pulling and then reading the data from http/websites
#beautifulSoup library is a web scraping libary to parse the document
#train_test_split for splitting data into train and test set
#confusion_matrix creates cross tab matrix between actual and predicted output

from sklearn.tree import DecisionTreeClassifier, export_graphviz

import pandas as pd
import gmaps

import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np 
from scipy.stats import norm 

from scipy import stats 
import requests 

from bs4 import BeautifulSoup  
import re
import json
from sklearn import tree
import pydotplus 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix

list1=[]
```

```python
#read URLs of different pages(2-30) from trulia website..
#Store all URLS of 30 pages in URLs list, which is passed as an object in function LoadData() in next block
URLs=["https://www.trulia.com/CA/San_Francisco"]
for i in range(2,30):
    URLs.append("https://www.trulia.com/CA/San_Francisco/{0}_p".format(i))
```

```python
#consider LoadData function a blackbox that converts a web page data of Housing posts into data frames, for each attribute.    
#To do so,Scrapping performs following operations
#1.)Calls requests.get() function that pings the trulia website for information
#2.)Convert the data into .txt format.
#3.)Calls beautiful soup function that parses the content in html_doc, converting the html_doc into unicode character 
#4.)Calls prettify which turns Beautiful Soup parse tree into a nicely formatted Unicode string, with each HTML tag on its own line. 
#5.)using find_all function, finds all tags with name "script" and save it in script2 variable
#6.)searches pageConfig word in "script2" variable,pageconfig contains all relevant housing data
#7.)Returns text as a single Unicode string
#8.)After removing the unnecessary words from the data, the final data is stored in fscript
#9.)this fscript data is written to an output.txt file
#10.) To convert the json data into dictionary readable format, the output.txt file is read and json.loads function is called
#11.) Finally truliaData contains all relevant data, in a nested dictionary format

def LoadData(url):
    r=requests.get(url)
    html_doc=r.text 
    soup=BeautifulSoup(html_doc)
    pretty_soup=soup.prettify()
    script2=soup.find_all("script")
    for i in range(len(script2)):
        if "pageConfig" in script2[i].get_text():
            script1=script2[i].get_text() 
    script=script1[18:]
    splitscript=script.split("var googleMapURL")
    fscript1=splitscript[0]
    fscript=fscript1[:-4]
    text_file = open("output.txt", "w")
    text_file.write(fscript)
    text_file.close()
    F=open("output.txt","rt")
    readtext=F.read()
    F.close()
    TruliaData1=json.loads(readtext)
    return TruliaData1
```

```python
#This block of code performs following operations
#1.)Creates list variables that stores data for corresponding attributes from website.
#2.)Creates a dictionary "parentlist", that contains all the UNIQUE keys from TrualiaData dictionary. 
#3.)TruliaData dictionary is a nested dictionary 
#4.)if keys in TruliaData["page"]["cards"] dictionary and parentlist dictionary matches, 
    #the list objects(area, city,price,bedroom etc) are populated with corresponding dictionary values 
#5.)Step 4 happens for each apartment post on each page, where each page has around 30 posts
#6.)For missing values specific to particular keys, Null is APPENDED to list object so as to keep same number of records

parentlist=dict()
stateCode=[]
OpenHouse=[]# categorical. 1 if openhouse, 0 otherwise
area=[]
city=[]
bedroom=[] #number of bedrooms in apartment
ID=[]
sqft=[] #variable named after square feet of apartment 
latLng=[] # containg latitude and longitude data
streetAddress=[] 
price=[]# price of an apartment
tags=[]
baths=[]#number of baths
Htags=[]

for t in range(len(URLs)):
    TruliaData=LoadData(URLs[t])
    for j in range(len(TruliaData["page"]["cards"])):
        for k in TruliaData["page"]["cards"][j].keys():
            if k not in parentlist:
                parentlist[k]=1
            else:
                parentlist[k]=parentlist[k]+1
                
for t in range(len(URLs)):
    TruliaData=LoadData(URLs[t])                
    for i in parentlist.keys():
        if i=='stateCode':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'stateCode' in TruliaData["page"]["cards"][j].keys():
                    stateCode.append(TruliaData["page"]["cards"][j]["stateCode"])
                else:
                    stateCode.append("Null")
        

        if i=='baths':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'baths' in TruliaData["page"]["cards"][j].keys():
                    baths.append(TruliaData["page"]["cards"][j]["baths"])
                else:
                    baths.append("Null")
        
        if i=='footer':
            for j in range(len(TruliaData["page"]["cards"])):
                area1=[]
                city1=[]
                alpha=TruliaData["page"]["cards"][j]["footer"]["location"].replace("'","")
                a=alpha.split(",")
                area1.append(a[0])
                city1.append(a[1])
                if 'footer' in TruliaData["page"]["cards"][j].keys():
                    area.append(a[0])
                    city.append(a[1])
                
                else:
                    area.append("Null")
                    city.append("Null")
        #print(temp,"And the length is",len(temp))   
        
   
        if i=='beds':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'beds' in TruliaData["page"]["cards"][j].keys():
                    bedroom.append(TruliaData["page"]["cards"][j]["beds"])
                else:
                    bedroom.append("Null")    
        if i=='id':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'id' in TruliaData["page"]["cards"][j].keys():
                    ID.append(TruliaData["page"]["cards"][j]["id"])
                else:
                    ID.append("Null")
        if i=='sqft':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'sqft' in TruliaData["page"]["cards"][j].keys():
                    sqft.append(TruliaData["page"]["cards"][j]["sqft"])
                else:
                    sqft.append("Null")
        
    
        if i=='latLng':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'latLng' in TruliaData["page"]["cards"][j].keys():
                    latLng.append(TruliaData["page"]["cards"][j]["latLng"])
                else:
                    latLng.append("Null")
        
        
        
        if i=='streetAddress':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'streetAddress' in TruliaData["page"]["cards"][j].keys():
                    streetAddress.append(TruliaData["page"]["cards"][j]["streetAddress"])
                else:
                    streetAddress.append("Null")
        
        
        if i=='price':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'price' in TruliaData["page"]["cards"][j].keys():
                    price.append(TruliaData["page"]["cards"][j]["price"])
                else:
                    price.append("Null")
        if i=='tags':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'tags' in TruliaData["page"]["cards"][j].keys():
                    tags.append(TruliaData["page"]["cards"][j]["tags"])
                else:
                    tags.append("Null")    
        
        if i=='highlightedTags':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'highlightedTags' in TruliaData["page"]["cards"][j].keys():
                    Htags.append(TruliaData["page"]["cards"][j]["highlightedTags"])
                else:
                    Htags.append("Null")
        
        
        if i=='hasOpenHouse':
            for j in range(len(TruliaData["page"]["cards"])):
                if 'hasOpenHouse' in TruliaData["page"]["cards"][j].keys():
                    OpenHouse.append(TruliaData["page"]["cards"][j]["hasOpenHouse"])
                else:
                    OpenHouse.append("Null") 



```

```python
#Data cleaning is performed here. 
# all special characters irrelevant to data are removed. 
#A final dictionary is created "TruliaDict"that joins all the list variables defined above into a single dictionary


for i in range(len(bedroom)):
    if bedroom[i].split("bd")[0]:
        bedroom[i]=bedroom[i].split("bd")[0]
    else:
        continue

for i in range(len(sqft)):
    sqft[i]=sqft[i].replace("sqft","")
    sqft[i]=sqft[i].replace(" sqft","")
    sqft[i]=sqft[i].replace(",","")

    
for i in range(len(price)):
    price[i]=price[i].replace("$","")
    price[i]=price[i].replace("+","")
    price[i]=price[i].replace(",","")
    price[i]=price[i].replace("'","")
    price[i]=price[i].replace(" ","")

TruliaDict=dict()
TruliaDict={"Htags":Htags,"OpenHouse":OpenHouse,"sqft":sqft,"tags":tags,"StateCode":stateCode,"area":area,"city":city,"bedroom":bedroom,"latLng":latLng,"streetAddress":streetAddress,"price":price}
TruliaDict.keys()    
```

```python
# Using Pandas, dictionary TruliaDict is converted to Dataframe data
#All cleaning and transformation are performed on this data
#replacing null with 0, for sqft, openhouse and bedroom variable
#converting sqft and bedroom to integer, using "astype" function
#converting area,latLng,OpenHouse to categorical,using "astype" function
#bedrooms range is reduced to 1-4,to avoid outliers with bedroom values like 8 or any higher number
#bedrooms range is reduced also to not have too many dummy variables for ir

#NEW VARIABLE Price/SQFR is created

data=pd.DataFrame(TruliaDict)
data.sqft.replace('Null',0,inplace=True)
data.OpenHouse.replace('Null',0,inplace=True)
data.OpenHouse.replace('True',1,inplace=True)
data.bedroom.replace('Null',0,inplace=True)
data['sqft']=data['sqft'].astype(int)
data["OpenHouse"]=data["OpenHouse"].astype('category')
data['price']=data['price'].astype(int)
data['bedroom'] = data['bedroom'].astype(int)
data['area'] = data['area'].astype('category')
#data['latLng'] = data['latLng'].astype(dtype)
data=data[data.sqft!=0] #to avoid infinity value in newly created variable (price/sqft)
data=data[data.bedroom<5]
data=data[data.bedroom>0]
data.info()

data['price/sqft']=data['price']/data['sqft']
data['bedroom'] = data['bedroom'].astype('category')
data['area'] = data['area'].astype('category')

data.info()
```

```python
#Since bedroom is a categorical variable,created 3 dummy variables for 4 bedroom apartment
#Dummy variables created using get_dummies function from pandas library.
# Using join function, joined all bedroom dummy variables with other variables in final dataframe "data"
#Drop na values from data
dummy_ranks = pd.get_dummies(data['bedroom'], prefix='bedrooms')
ImportantCol = ['area', 'price','OpenHouse','sqft','bedroom','price/sqft','latLng']
data = data[ImportantCol].join(dummy_ranks.ix[:, 'bedrooms_2':])
data=data.dropna()
data
```

```python
###Exploratory Data Analysis part 1
#Since the price/sqft wasn't linearly distributed, transformed the price/sqft to log(price/sqft) 
#As seen in the distplot, log(sqft/price) is normally distributed 
#As seen in the probplot(like a Q-Q plot),log(sqft/price) points fall in line  
#set function sets the figsize according to the need 

sns.set(font_scale=1,rc={"figure.figsize": (5,5)}); 
data['price/sqft'] = np.log(data['price/sqft'])
sns.distplot(data['price/sqft'], fit=norm);
fig = plt.figure()
fplot = stats.probplot(data['price/sqft'], plot=plt)

```

```python
# X variable is a set of independent variables(IVs) to be used in decision tree model
# Y is the dependent/outcome variable containing  variable "area" from dataframe data

tempvar=['price/sqft','sqft','bedrooms_2','bedrooms_3','bedrooms_4']
X=data[tempvar]
tempvar2=['area']
Y=data[tempvar2]
Y['area']=Y['area'].astype('category')
X['price/sqft']=X['price/sqft'].astype('category')

X['sqft']=X['sqft'].astype('category')
X['bedrooms_2']=X['bedrooms_2'].astype('category')
X['bedrooms_3']=X['bedrooms_3'].astype('category')
X['bedrooms_4']=X['bedrooms_4'].astype('category')
```

```python
# Data is now split into train and test datasets.. 
# training and testing data sets are 80% and 20% respectively of the total data
X_train, X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```python
# Decision tree classifier function is used to run the decision tree
#After fitting, the variables are ranked with importance they play in decision tree
# column sqft has the highest importance score of 0.484, which means this column is most significant 
#dectree.score tells us how good is the model. With score=.11, this model is a poor predictor of Y_test(area in san francisco) 
#
DT = tree.DecisionTreeClassifier()
dectree=DT.fit(X_train,Y_train)
print("['price/sqft','sqft','bedrooms_2','bedrooms_3','bedrooms_4']")
print(dectree.feature_importances_)
print(dectree.score(X_test,Y_test))
```

```python
#After being fitted, the model can be used to predict the class of samples
#Areatrain=Y_test["area"].values
#Consufion matrix has: 
#1.) The diagonal elements which show number of correct classifications for each class: 
#2.) The off-diagonal elements which show the misclassifications

temp1=dectree.predict_proba(X_test)
alpha1=dectree.predict(X_test)
print(confusion_matrix(Y_test,alpha1))
X_test["Predicted Output"]=alpha1
FinalData=X_test
FinalData
```
